```python
#第五步
#对gostar进行本身去重
#逐行搜索一个csv文件，根据每行的第一、第五、第七列生成的[a,b]对数据进行去重，如果有重复的[a,b]则将整行删去，保存不重复的行为新的csv
import csv

def remove_duplicates(input_file, output_file):
    data = set()  # 用于存储已经出现过的[a,b,c]

    with open(input_file, 'r', newline='') as file:
        reader = csv.reader(file)
        #reader = csv.reader(itertools.islice(file, 200))
        rows = list(reader)

        unique_rows = []
        for row in rows:
            a, b, c = row[8], row[9], row[10]
            #print((a, b, c))
            if (a, b, c) not in data:
                data.add((a, b, c))
                unique_rows.append(row)

    with open(output_file, 'w', newline='') as file:
        writer = csv.writer(file)
        writer.writerows(unique_rows)

# 替换为实际的文件路径
input_file = '/mnt/d/gostar/0或1对比后单独结果.csv'
output_file = '/mnt/d/gostar/0或1对比后单独结果_添加本身去重.csv'
remove_duplicates(input_file, output_file)
```

```python
#第六步
#对gostar进行pdb去重
#你可以使用以下 Python 代码逐行搜索一个 CSV 文件的第一列，并检查每行的字符是否存在于另一个 CSV 文件的第三列中。如果字符存在，则保留整个行；否则，删除整行。
import csv

def filter_rows(input_file, filter_file, output_file):
    filter_data = set()  # 存储第三列的数据

    # 读取过滤文件的第三列数据
    with open(filter_file, 'r', newline='') as filter_csv:
        filter_reader = csv.reader(filter_csv)
        next(filter_reader)  # 跳过标题行
        for row in filter_reader:
            filter_data.add(row[2])

    # 过滤并保留包含在过滤数据中的行
    filtered_rows = []
    with open(input_file, 'r', newline='') as input_csv:
        input_reader = csv.reader(input_csv)
        for row in input_reader:
            if row[8] in filter_data:
                filtered_rows.append(row)

    # 将结果写入输出文件
    with open(output_file, 'w', newline='') as output_csv:
        writer = csv.writer(output_csv)
        writer.writerows(filtered_rows)

# 替换为实际的文件路径
input_file = '/mnt/d/gostar/0或1对比后单独结果_添加本身去重.csv' # 要处理的CSV文件
filter_file = '/mnt/d/gostar/protein_data.csv'  # 用于过滤的CSV文件
output_file = '/mnt/d/gostar/0或1对比后单独结果_添加本身去重_添加pdb去重.csv'  # 结果输出的CSV文件

filter_rows(input_file, filter_file, output_file)
```

```python
#第七步
########################### pdb搜索

import csv

input_file = '/mnt/d/gostar/0或1对比后单独结果_添加本身去重_添加pdb去重.csv'   # 第一列字符CSV文件路径
search_file = '/mnt/d/gostar/protein_data.csv'  # 搜索CSV文件路径
output_file = '/mnt/d/gostar/0或1对比后单独结果_添加本身去重_添加pdb去重_添加pdb值.csv'   # 输出CSV文件路径

with open(input_file, 'r') as f_input, open(search_file, 'r') as f_search, open(output_file, 'w', newline='') as f_output:
    input_reader = csv.reader(f_input)
    search_reader = csv.reader(f_search)
    output_writer = csv.writer(f_output)

    # 读取search文件的第一列和第二列的值生成一个字典
    search_dict = {}
    for search_row in search_reader:
        search_dict[search_row[2]] = search_row[0]

    # 处理input文件中的每一行
    for input_row in input_reader:
        character = input_row[8]

        if character in search_dict:
            value = search_dict[character]
            input_row.append(value)

        output_writer.writerow(input_row)

```

```python
#第八步
#############################移动pdb
import os
import shutil
import csv

# 读取CSV文件的路径和目标文件夹路径
csv_file = '/mnt/d/gostar/0或1对比后单独结果_添加本身去重_添加pdb去重_添加pdb值.csv' 
#csv_file = '/mnt/d/gostar/信息处理（指定口袋对接）（带pdb正序）.csv'
source_folder = '/mnt/d/gostar/PDBbind_v2020_other_PL.tar/PDBbind_v2020_other_PL/v2020-other-PL'
#source_folder = '/mnt/d/gostar/PDBbind_v2020_other_PL.tar/PDBbind_v2020_other_PL/refined-set'
destination_folder = '/mnt/f/pdb_all_dayudengyu'

# 读取CSV文件的第一列和第二列的每一行字符
characters = []
new_names = []
with open(csv_file, 'r') as f:
    reader = csv.reader(f)
    for row in reader:
        character = row[11]
        new_name = row[8].replace(' ', '_')
        characters.append(character)
        new_names.append(new_name)

# 在源文件夹中查找与字符名称相同的文件夹，并将其复制并重命名到目标文件夹位置
for character, new_name in zip(characters, new_names):
    source_path = os.path.join(source_folder, character)
    destination_path = os.path.join(destination_folder, new_name)

    try:
        if os.path.exists(source_path) and os.path.isdir(source_path) and not os.path.exists(destination_path):
            shutil.copytree(source_path, destination_path)
    except FileExistsError:
        continue
```

```python
#第9步
#统计smiles
import csv

def count_duplicates(csv_file):
    counts = {}  # 用于记录每个字符串出现次数的字典

    with open(csv_file, 'r') as file:
        reader = csv.reader(file)
        rows = list(reader)  # 将CSV文件读取为行列表

    with open(csv_file, 'w', newline='') as file:
        writer = csv.writer(file)

        for row in rows:
            string = row[8]  # 获取当前行的第一个元素，即要搜索的字符串

            if string not in counts:
                counts[string] = 1
            else:
                counts[string] += 1

            row.append(counts[string])  # 在当前行的最后一列添加计数结果
            writer.writerow(row)

# 示例使用
csv_file = '/mnt/d/gostar/0或1对比后单独结果_添加本身去重_添加pdb去重_添加pdb值.csv'
count_duplicates(csv_file)
```

```python
#第10步
#sdf文件制作
import os
import csv
from openbabel import pybel

def process_folders(folder_path, csv_file):
    if not os.path.exists(folder_path):
        print("文件夹路径不存在")
        return

    if not os.path.isfile(csv_file):
        print("CSV文件路径无效")
        return

    with open(csv_file, 'r') as file:
        reader = csv.reader(file)
        csv_data = list(reader)

    for folder_name in os.listdir(folder_path):
        folder_dir = os.path.join(folder_path, folder_name)
        if os.path.isdir(folder_dir):
            for row in csv_data:
                if row[9] == folder_name:
                    smiles = row[9]
                    output_filename = row[12]
                    if smiles != '':
                        mol = pybel.readstring('smi', smiles)
                        mol.addh()
                        mol.make3D()
                    else:  
                        print("Empty smiles string")  
                    if mol is not None:
                        dock_input_folder = os.path.join(folder_dir, "docking_input")
                        if not os.path.exists(dock_input_folder):
                            os.makedirs(dock_input_folder)
                        output_file_path = os.path.join(dock_input_folder, f"{output_filename}.sdf")
                        mol.write('sdf', output_file_path, overwrite=True)
                        print(f"已为文件夹 {folder_name} 创建了 docking_input/{output_filename}.sdf 文件")

# 示例使用
folder_path = '/mnt/f/pdb_all_dayudengyu' # 替换为要遍历的文件夹的路径
#csv_file = '/mnt/d/gostar/排序试验.csv' 
csv_file = '/mnt/d/gostar/0或1对比后单独结果_添加本身去重_添加pdb去重_添加pdb值_添加smiles序号.csv' # 替换为包含CSV数据的文件路径 
process_folders(folder_path, csv_file)
```

```python
#第十步蛋白水和离子去除
import os
from pymol import cmd

def remove_water_ions_metals_recursive(root_dir, output_dir):
    # 创建输出目录（如果不存在）
    if not os.path.exists(output_dir):
        os.makedirs(output_dir)

    # 遍历根目录下的所有子目录
    for dirpath, dirnames, filenames in os.walk(root_dir):
        # 遍历当前子目录中的文件
        for filename in filenames:
            if filename.endswith('.pdb'):
                # 构建当前 PDB 文件的完整路径
                pdb_path = os.path.join(dirpath, filename)

                # 打开 PyMOL
                cmd.reinitialize()

                # 加载当前 PDB 文件
                cmd.load(pdb_path)

                # 去除水分子
                cmd.remove('solvent')

                # 去除金属离子
                metals = ['NA+', 'CL-', 'ZN', 'FE', 'CU']
                for metal in metals:
                    cmd.remove(f'resn {metal}')

                # 提取纯净的蛋白质结构
                cmd.select('protein', 'polymer.protein')

                # 构建输出文件的路径和文件名
                output_subdir = os.path.join(output_dir, os.path.relpath(dirpath, root_dir))
                output_file = os.path.join(output_subdir, filename)

                # 创建输出子目录（如果不存在）
                if not os.path.exists(output_subdir):
                    os.makedirs(output_subdir)

                # 保存纯净的蛋白质结构
                cmd.save(output_file, 'protein')

# 替换为实际的目录路径
root_dir = '/mnt/f/pdb_all_dayudengyu' # 包含多个子文件夹的根目录
output_dir =  '/mnt/f/pdb_all_dayudengyu'  # 存储结果的目录

remove_water_ions_metals_recursive(root_dir, output_dir)
```
