```python
###########################在pdbbind里面进行pdb搜索

import csv

input_file = '/mnt/d/gostar/信息处理（指定口袋对接）.csv'  # 第一列字符CSV文件路径
search_file = '/mnt/d/gostar/protein_data.csv'  # 搜索CSV文件路径
output_file = '信息处理（指定口袋对接）（带pdb正序）.csv'  # 输出CSV文件路径

with open(input_file, 'r') as f_input, open(search_file, 'r') as f_search, open(output_file, 'w', newline='') as f_output:
    input_reader = csv.reader(f_input)
    search_reader = csv.reader(f_search)
    output_writer = csv.writer(f_output)

    # 读取search文件的第一列和第二列的值生成一个字典
    search_dict = {}
    for search_row in search_reader:
        search_dict[search_row[2]] = search_row[0]

    # 处理input文件中的每一行
    for input_row in input_reader:
        character = input_row[0]

        if character in search_dict:
            value = search_dict[character]
            input_row.append(value)

        output_writer.writerow(input_row)

```

```python
############################对搜索结果进行移动pdb
import os
import shutil
import csv

# 读取CSV文件的路径和目标文件夹路径
#csv_file = '/mnt/d/gostar/筛选结果18正序.csv'
csv_file = '/mnt/d/gostar/信息处理（指定口袋对接）（带pdb正序）.csv'
source_folder = '/mnt/d/gostar/PDBbind_v2020_other_PL.tar/PDBbind_v2020_other_PL/v2020-other-PL'
#source_folder = '/mnt/d/gostar/PDBbind_v2020_other_PL.tar/PDBbind_v2020_other_PL/refined-set'
destination_folder = '/mnt/f/pdb_all_18_zhengxu'

# 读取CSV文件的第一列和第二列的每一行字符
characters = []
new_names = []
with open(csv_file, 'r') as f:
    reader = csv.reader(f)
    for row in reader:
        character = row[2]
        new_name = row[0].replace(' ', '_')
        characters.append(character)
        new_names.append(new_name)

# 在源文件夹中查找与字符名称相同的文件夹，并将其复制并重命名到目标文件夹位置
for character, new_name in zip(characters, new_names):
    source_path = os.path.join(source_folder, character)
    destination_path = os.path.join(destination_folder, new_name)

    try:
        if os.path.exists(source_path) and os.path.isdir(source_path) and not os.path.exists(destination_path):
            shutil.copytree(source_path, destination_path)
    except FileExistsError:
        continue
```

```python
##搜索多个csv文件
import csv

# 第一个CSV文件路径
csv_file1 =  '/mnt/d/gostar/多pdb搜索.csv'
# 第二个CSV文件路径
csv_file2 = '/mnt/d/gostar/INDEX_refined_name2020.csv'

output_file = 'output.csv'

# 读取第二个CSV文件的第三列的值
lookup_values = set()
with open(csv_file2, 'r') as f2:
    reader2 = csv.reader(f2)
    for row2 in reader2:
        lookup_values.add(row2[2])

# 对第一个CSV文件进行搜索并复制值到最后一列
with open(csv_file1, 'r') as f1, open(output_file, 'w', newline='') as f_out:
    reader1 = csv.reader(f1)
    writer_out = csv.writer(f_out)
    header1 = next(reader1)
    writer_out.writerow(header1)
    
    for row1 in reader1:
        value_to_search = row1[0]
        
        if value_to_search in lookup_values:
            writer_out.writerow(row1)
```

```python
#smiles作用值搜索  9.19工作1
import csv

def copy_rows(csv_file1, csv_file2):
    rows1 = []  # 存储第一个 CSV 文件的行数据
    rows2 = []  # 存储第二个 CSV 文件的行数据

    with open(csv_file1, 'r') as file1:
        reader1 = csv.reader(file1)
        rows1 = list(reader1)

    with open(csv_file2, 'r') as file2:
        reader2 = csv.reader(file2)
        rows2 = list(reader2)

    for index, row1 in enumerate(rows1):
        a = row1[0]
        b = row1[1]
        matching_rows = []

        for row2 in rows2:
            c = row2[7]
            d = row2[8]

            if [a, b] == [c, d]:
                matching_rows.append(row2)  # 将匹配到的行保存在匹配行列表中

        if matching_rows:
            for matching_row in matching_rows:
                extended_row = row1 + matching_row  # 将匹配行数据与当前行进行合并
                rows1[index] = extended_row  # 替换当前行为合并后的行数据

    with open(csv_file1, 'w', newline='') as file1:
        writer = csv.writer(file1)
        writer.writerows(rows1)

# 示例使用
csv_file1 = '/mnt/d/gostar/信息处理（指定口袋对接）（带pdb正序）.csv'
#csv_file1 = '/mnt/d/gostar/用于smile_to_mol的排序文件'

#csv_file2 =  '/mnt/d/gostar/smilestry2.csv'
csv_file2 =  '/mnt/d/gostar/smiles值对应文件.csv'
copy_rows(csv_file1, csv_file2)
```

```python
#smiles计数
import csv

def count_duplicates(csv_file):
    counts = {}  # 用于记录每个字符串出现次数的字典

    with open(csv_file, 'r') as file:
        reader = csv.reader(file)
        rows = list(reader)  # 将CSV文件读取为行列表

    with open(csv_file, 'w', newline='') as file:
        writer = csv.writer(file)

        for row in rows:
            string = row[0]  # 获取当前行的第一个元素，即要搜索的字符串

            if string not in counts:
                counts[string] = 1
            else:
                counts[string] += 1

            row.append(counts[string])  # 在当前行的最后一列添加计数结果
            writer.writerow(row)

# 示例使用
csv_file = '/mnt/d/gostar/筛选结果 - 副本.csv'

count_duplicates(csv_file)
```

```python
#pdb在pbdbind里面有多少一对多的蛋白数量搜索
import csv

# 读取第一个 CSV 文件
csv_file_1 = '/mnt/d/gostar/dayudeengyujieguo.csv'
column_8_set = set()

with open(csv_file_1, 'r') as file:
    reader = csv.reader(file)
    
    for row in reader:
        if len(row) >= 8:
            column_8_set.add(row[8])  # 第八列的索引为 7，因为索引从 0 开始

# 读取第二个 CSV 文件
csv_file_2 = '/mnt/d/gostar/protein_data.csv'
element_count = {}  # 存储元素及其出现次数的字典

with open(csv_file_2, 'r') as file:
    reader = csv.reader(file)
    
    for row in reader:
        if len(row) >= 3 and row[2] in column_8_set:  # 第三列的索引为 2
            element = row[2]
            if element in element_count:
                element_count[element] += 1
            else:
                element_count[element] = 1

# 将结果写入输出 CSV 文件
output_file =  '/mnt/d/gostar/cishu.csv'

with open(output_file, 'w', newline='') as file:
    writer = csv.writer(file)
    writer.writerow(['ele', 'num'])  # 写入表头
    
    for element, count in element_count.items():
        writer.writerow([element, count])

print("结果已写入", output_file)
```
